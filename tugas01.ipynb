```python
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
original = data.astronaut()
print(original.shape)

plt.imshow(original)
```

```python
red = original[:, :, 0]
green = original[:, :, 1]
blue = original[:, :, 2]

print("Red shape: ", red.shape)
print("Green shape: ", green.shape)
print("Blue shape: ", blue.shape)

fig, axes = plt.subplots(2 , 2 , figsize=(12,12))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Citra Input")

ax[1].imshow(red, cmap=plt.cm.gray)
ax[1].set_title("Red Chanel")

ax[2].imshow(green, cmap=plt.cm.gray)
ax[2].set_title("Green Chanel")

ax[3].imshow(blue, cmap=plt.cm.gray)
ax[3].set_title("Blue Chanel")

fig.tight_layout()
plt.show()
```

```python
red = original.copy()
red[:, :, 1]=0
red[:, :, 2]=0
print("Red Shape= ", red.shape)

green = original.copy()
green[:, :, 0]=0
green[:, :, 2]=0
print("Green Shape= ", green.shape)

blue = original.copy()
blue[:, :, 0]=0
blue[:, :, 1]=0
print("Blue Shape= ", blue.shape)

fig, axes = plt.subplots(2 , 2 , figsize=(12,12))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Citra Input")

ax[1].imshow(red, cmap=plt.cm.gray)
ax[1].set_title("Red Chanel")

ax[2].imshow(green, cmap=plt.cm.gray)
ax[2].set_title("Green Chanel")

ax[3].imshow(blue, cmap=plt.cm.gray)
ax[3].set_title("Blue Chanel")

fig.tight_layout()
plt.show()
```
